### Fichier d'analyse des résulats

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv', sep=' ;', index=False)
print(df.columns)
```

```python
print(sum(df["Nucléaire"]))
print(sum(df["STEP turbinage"]))
```

```python
plt.figure(figsize=(11,5))

plt.stackplot(
    df.index,
    [df[c] for c in df.columns],
    labels=df.columns
)

plt.plot(df.index, df.sum(axis=1), color="black", linewidth=1.8, label="Production totale")

plt.title("Production électrique par filière (empilée)")
plt.xlabel("Temps")
plt.ylabel("Puissance (MW)")
plt.legend(ncol=2)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
